## 2. Unificação dos Dados

```python
# Lista de todos os DataFrames
df = {
    2014: df_2014, 2015: df_2015, 2016: df_2016, 2017: df_2017,
    2018: df_2018, 2019: df_2019, 2020: df_2020, 2021: df_2021,
    2022: df_2022, 2023: df_2023
}
```

```python
# Lista para armazenar os dados processados de cada ano
yearly_data = []

for year, df in df.items():
    # Extrai o sufixo do ano (ex: 14, 15, 23)
    year_suffix = str(year)[-2:]

    # Seleciona apenas as colunas relevantes (cnae e aquelas que terminam com o sufixo do ano)
    cols_to_keep = ['cnae'] + [col for col in df.columns if col.endswith(f'_{year_suffix}')]

    df_year = df[cols_to_keep].copy()

    # Renomeia as colunas
    new_column_names = {
        'cnae': 'cnae',
        f'tot_{year_suffix}': 'total_acidentes',
        f'tcat_{year_suffix}': 'com_cat_registrada',
        f'tip_{year_suffix}': 'acidentes_tipicos',
        f'trj_{year_suffix}': 'acidentes_trajeto',
        f'doe_{year_suffix}': 'doenca_trabalho',
        f'scat_{year_suffix}': 'sem_cat_registrada'
    }
    df_year.rename(columns=new_column_names, inplace=True)

    # Adiciona a coluna 'ano'
    df_year['ano'] = year

    yearly_data.append(df_year)

# Concatena todos os dataframes anuais em um só
df_consolidado = pd.concat(yearly_data, ignore_index=True)

print("Formato do DataFrame consolidado:", df_consolidado.shape)
df_consolidado.head()
```

## 4. Limpeza e Conversão de Tipos

```python
# Seleciona as colunas de contagem de acidentes
colunas_numericas = [
    'total_acidentes', 'com_cat_registrada', 'acidentes_tipicos',
    'acidentes_trajeto', 'doenca_trabalho', 'sem_cat_registrada'
]

# Limpa e converte
for col in colunas_numericas:
    # Garante que a coluna seja do tipo string para usar o .str
    df_consolidado[col] = df_consolidado[col].astype(str)

    # Substitui os caracteres não-numéricos
    df_consolidado[col] = df_consolidado[col].str.replace('.', '', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('–', '0', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('-', '0', regex=False)

    # Substitui a string 'nan'
    df_consolidado[col] = df_consolidado[col].str.replace('nan', '0', regex=False)

    # Converte para numérico, preenchendo valores nulos com 0
    df_consolidado[col] = pd.to_numeric(df_consolidado[col]).fillna(0).astype(int)

# Garante que 'cnae' seja tratado como string
df_consolidado['cnae'] = df_consolidado['cnae'].astype(str)

# Tipo dos dados do novo df
print(df_consolidado.info())
```

## 5. Filtragem do Setor Petroquímico
CNAE : ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

```python
# Define os CNAEs específicos (primeiros 4 dígitos)
cnaes_especificos = ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

# Filtra os registros CNAE que começam com os códigos específicos
df_petroquimico = df_consolidado[df_consolidado['cnae'].isin(cnaes_especificos)].copy().reset_index(drop=True)
df_petroquimico.head()
```

```python
# Ordena os dados por cnae e ano
df_petroquimico = df_petroquimico.sort_values(by=['cnae', 'ano']).reset_index(drop=True)
df_petroquimico.head()
```

## 6. Feature Engineering
Definição de Múltiplos Targets

```python
# Define a lista de colunas que queremos prever
colunas_alvo = [
    'total_acidentes',
    'com_cat_registrada',
    'acidentes_tipicos',
    'acidentes_trajeto',
    'doenca_trabalho',
    'sem_cat_registrada'
]

# Cria um df temporário com os targets
df_alvos = df_petroquimico[['cnae'] + colunas_alvo]
df_alvos.head()
```

```python
# Cria outro df com os valores das colunas do ano seguinte para serem os targets
y_multi_alvo = df_alvos.groupby('cnae').shift(-1)
y_multi_alvo.head()
```

```python
# Concatena os dois dfs e remove linhas com NaN
df_final_multi = pd.concat([df_petroquimico, y_multi_alvo.add_suffix('_seguinte')], axis=1)
df_final_multi = df_final_multi.dropna()
df_final_multi.head()
```

```python
# Importação das bibliotecas de visualização
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurações de estilo
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("Bibliotecas de visualização importadas com sucesso!")
```

### 9.1. Evolução Temporal dos Acidentes por Ano

```python
# Agrupa dados por ano e soma os acidentes
acidentes_por_ano = df_petroquimico.groupby('ano').agg({
    'total_acidentes': 'sum',
    'acidentes_tipicos': 'sum',
    'acidentes_trajeto': 'sum',
    'doenca_trabalho': 'sum'
}).reset_index()

# Cria o gráfico
fig, ax = plt.subplots(figsize=(14, 7))

# Plota as linhas
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['total_acidentes'], 
        marker='o', linewidth=2.5, markersize=8, label='Total de Acidentes', color='#d62728')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['acidentes_tipicos'], 
        marker='s', linewidth=2, markersize=7, label='Acidentes Típicos', color='#1f77b4')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['acidentes_trajeto'], 
        marker='^', linewidth=2, markersize=7, label='Acidentes de Trajeto', color='#ff7f0e')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['doenca_trabalho'], 
        marker='D', linewidth=2, markersize=7, label='Doenças do Trabalho', color='#2ca02c')

# Personalização
ax.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax.set_ylabel('Número de Acidentes', fontsize=12, fontweight='bold')
ax.set_title('Evolução Temporal dos Acidentes de Trabalho no Setor Petroquímico (2014-2023)', 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='best', fontsize=10, framealpha=0.9)
ax.grid(True, alpha=0.3)
ax.set_xticks(acidentes_por_ano['ano'])

# Adiciona valores no topo de cada ponto
for i, row in acidentes_por_ano.iterrows():
    ax.text(row['ano'], row['total_acidentes'] + 300, f"{int(row['total_acidentes']):,}", 
            ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# Exibe estatísticas
print(f"\n{'='*60}")
print(f"ESTATÍSTICAS DE EVOLUÇÃO TEMPORAL")
print(f"{'='*60}")
print(f"Ano com MAIS acidentes: {acidentes_por_ano.loc[acidentes_por_ano['total_acidentes'].idxmax(), 'ano']:.0f} ({acidentes_por_ano['total_acidentes'].max():,} acidentes)")
print(f"Ano com MENOS acidentes: {acidentes_por_ano.loc[acidentes_por_ano['total_acidentes'].idxmin(), 'ano']:.0f} ({acidentes_por_ano['total_acidentes'].min():,} acidentes)")
print(f"Variação 2014-2023: {((acidentes_por_ano.iloc[-1]['total_acidentes'] / acidentes_por_ano.iloc[0]['total_acidentes']) - 1) * 100:.1f}%")
print(f"{'='*60}\n")
```

### 9.2. Distribuição por Tipo de Acidente

```python
# Calcula a distribuição total por tipo de acidente
distribuicao_tipos = {
    'Acidentes Típicos': df_petroquimico['acidentes_tipicos'].sum(),
    'Acidentes de Trajeto': df_petroquimico['acidentes_trajeto'].sum(),
    'Doenças do Trabalho': df_petroquimico['doenca_trabalho'].sum()
}

# Cria subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico de Pizza
cores = ['#1f77b4', '#ff7f0e', '#2ca02c']
wedges, texts, autotexts = ax1.pie(distribuicao_tipos.values(), 
                                     labels=distribuicao_tipos.keys(),
                                     autopct='%1.1f%%',
                                     colors=cores,
                                     startangle=90,
                                     explode=(0.05, 0.05, 0.05),
                                     shadow=True)

# Personaliza os textos do gráfico de pizza
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(11)
    autotext.set_fontweight('bold')
    
for text in texts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

ax1.set_title('Distribuição Proporcional dos Tipos de Acidentes\n(2014-2023)', 
              fontsize=13, fontweight='bold', pad=20)

# Gráfico de Barras Horizontais
tipos = list(distribuicao_tipos.keys())
valores = list(distribuicao_tipos.values())
barras = ax2.barh(tipos, valores, color=cores, edgecolor='black', linewidth=1.5)

# Adiciona valores nas barras
for i, (bar, valor) in enumerate(zip(barras, valores)):
    ax2.text(valor + 500, i, f'{valor:,}', va='center', fontsize=11, fontweight='bold')

ax2.set_xlabel('Número Total de Acidentes', fontsize=12, fontweight='bold')
ax2.set_title('Quantidade Total por Tipo de Acidente\n(2014-2023)', 
              fontsize=13, fontweight='bold', pad=20)
ax2.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Exibe estatísticas
print(f"\n{'='*60}")
print(f"DISTRIBUIÇÃO POR TIPO DE ACIDENTE (2014-2023)")
print(f"{'='*60}")
total = sum(distribuicao_tipos.values())
for tipo, qtd in distribuicao_tipos.items():
    percentual = (qtd / total) * 100
    print(f"{tipo:.<30} {qtd:>7,} ({percentual:>5.1f}%)")
print(f"{'='*60}")
print(f"{'TOTAL':.<30} {total:>7,} (100.0%)")
print(f"{'='*60}\n")
```

### 9.3. Comparativo de Acidentes por CNAE

```python
# Dicionário com descrições dos CNAEs
cnae_descricoes = {
    '0910': 'Extração de Petróleo e Gás',
    '1921': 'Fabricação de Produtos Petroquímicos',
    '2513': 'Embalagens Metálicas',
    '3321': 'Instalação de Máquinas',
    '3520': 'Distribuição de Gás',
    '4223': 'Obras de Infraestrutura',
    '4292': 'Montagem de Estruturas',
    '4321': 'Instalações Elétricas',
    '4940': 'Transporte Dutoviário',
    '7112': 'Serviços de Engenharia'
}

# Agrupa por CNAE
acidentes_por_cnae = df_petroquimico.groupby('cnae').agg({
    'total_acidentes': 'sum',
    'acidentes_tipicos': 'sum',
    'acidentes_trajeto': 'sum',
    'doenca_trabalho': 'sum'
}).reset_index()

# Ordena por total de acidentes
acidentes_por_cnae = acidentes_por_cnae.sort_values('total_acidentes', ascending=True)

# Adiciona descrições
acidentes_por_cnae['descricao'] = acidentes_por_cnae['cnae'].map(cnae_descricoes)
acidentes_por_cnae['label'] = acidentes_por_cnae['cnae'] + ' - ' + acidentes_por_cnae['descricao']

# Cria o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(14, 8))

y_pos = np.arange(len(acidentes_por_cnae))
width = 0.8

# Barras empilhadas
p1 = ax.barh(y_pos, acidentes_por_cnae['acidentes_tipicos'], width, 
             label='Acidentes Típicos', color='#1f77b4', edgecolor='black', linewidth=0.5)
p2 = ax.barh(y_pos, acidentes_por_cnae['acidentes_trajeto'], width,
             left=acidentes_por_cnae['acidentes_tipicos'],
             label='Acidentes de Trajeto', color='#ff7f0e', edgecolor='black', linewidth=0.5)
p3 = ax.barh(y_pos, acidentes_por_cnae['doenca_trabalho'], width,
             left=acidentes_por_cnae['acidentes_tipicos'] + acidentes_por_cnae['acidentes_trajeto'],
             label='Doenças do Trabalho', color='#2ca02c', edgecolor='black', linewidth=0.5)

# Adiciona o total no final de cada barra
for i, (idx, row) in enumerate(acidentes_por_cnae.iterrows()):
    ax.text(row['total_acidentes'] + 500, i, f"{int(row['total_acidentes']):,}", 
            va='center', fontsize=10, fontweight='bold')

# Personalização
ax.set_yticks(y_pos)
ax.set_yticklabels(acidentes_por_cnae['label'], fontsize=10)
ax.set_xlabel('Número de Acidentes', fontsize=12, fontweight='bold')
ax.set_title('Comparativo de Acidentes por CNAE no Setor Petroquímico (2014-2023)', 
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='lower right', fontsize=10, framealpha=0.9)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

# Exibe estatísticas
print(f"\n{'='*80}")
print(f"TOP 5 CNAEs COM MAIS ACIDENTES (2014-2023)")
print(f"{'='*80}")
top5 = acidentes_por_cnae.sort_values('total_acidentes', ascending=False).head()
for idx, row in top5.iterrows():
    print(f"{row['cnae']} - {row['descricao']:<40} {int(row['total_acidentes']):>7,} acidentes")
print(f"{'='*80}\n")
```

### 9.4. Proporção de Acidentes com e sem CAT Registrada

```python
# Agrupa por ano para ver evolução da CAT
cat_por_ano = df_petroquimico.groupby('ano').agg({
    'com_cat_registrada': 'sum',
    'sem_cat_registrada': 'sum',
    'total_acidentes': 'sum'
}).reset_index()

# Cria subplots
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)

# 1. Gráfico de linha - Evolução temporal
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(cat_por_ano['ano'], cat_por_ano['com_cat_registrada'], 
         marker='o', linewidth=2.5, markersize=8, label='Com CAT Registrada', color='#2ca02c')
ax1.plot(cat_por_ano['ano'], cat_por_ano['sem_cat_registrada'], 
         marker='s', linewidth=2.5, markersize=8, label='Sem CAT Registrada', color='#d62728')
ax1.fill_between(cat_por_ano['ano'], cat_por_ano['com_cat_registrada'], alpha=0.3, color='#2ca02c')
ax1.fill_between(cat_por_ano['ano'], cat_por_ano['sem_cat_registrada'], alpha=0.3, color='#d62728')

ax1.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax1.set_ylabel('Número de Acidentes', fontsize=12, fontweight='bold')
ax1.set_title('Evolução Temporal: Acidentes com vs sem CAT Registrada', 
              fontsize=13, fontweight='bold', pad=15)
ax1.legend(loc='best', fontsize=11)
ax1.grid(True, alpha=0.3)
ax1.set_xticks(cat_por_ano['ano'])

# 2. Gráfico de pizza - Proporção total
ax2 = fig.add_subplot(gs[1, 0])
total_com_cat = df_petroquimico['com_cat_registrada'].sum()
total_sem_cat = df_petroquimico['sem_cat_registrada'].sum()

cores_cat = ['#2ca02c', '#d62728']
wedges, texts, autotexts = ax2.pie([total_com_cat, total_sem_cat], 
                                     labels=['Com CAT\nRegistrada', 'Sem CAT\nRegistrada'],
                                     autopct='%1.1f%%',
                                     colors=cores_cat,
                                     startangle=90,
                                     explode=(0.05, 0.05),
                                     shadow=True)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')
    
for text in texts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

ax2.set_title('Proporção Total (2014-2023)', fontsize=12, fontweight='bold', pad=15)

# 3. Taxa de registro por ano
ax3 = fig.add_subplot(gs[1, 1])
cat_por_ano['taxa_registro'] = (cat_por_ano['com_cat_registrada'] / cat_por_ano['total_acidentes']) * 100

barras = ax3.bar(cat_por_ano['ano'], cat_por_ano['taxa_registro'], 
                 color='#17becf', edgecolor='black', linewidth=1.5, alpha=0.8)

# Adiciona valores nas barras
for bar in barras:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

ax3.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax3.set_ylabel('Taxa de Registro (%)', fontsize=12, fontweight='bold')
ax3.set_title('Taxa de Registro CAT por Ano', fontsize=12, fontweight='bold', pad=15)
ax3.set_xticks(cat_por_ano['ano'])
ax3.set_xticklabels(cat_por_ano['ano'], rotation=45)
ax3.grid(axis='y', alpha=0.3)
ax3.set_ylim(0, 100)

# Adiciona linha de referência
ax3.axhline(y=cat_por_ano['taxa_registro'].mean(), color='red', 
            linestyle='--', linewidth=2, label=f'Média: {cat_por_ano["taxa_registro"].mean():.1f}%')
ax3.legend(fontsize=10)

plt.tight_layout()
plt.show()

# Exibe estatísticas
print(f"\n{'='*70}")
print(f"ESTATÍSTICAS DE REGISTRO CAT (2014-2023)")
print(f"{'='*70}")
print(f"Com CAT Registrada.............: {total_com_cat:>10,} ({(total_com_cat/(total_com_cat+total_sem_cat)*100):>5.1f}%)")
print(f"Sem CAT Registrada.............: {total_sem_cat:>10,} ({(total_sem_cat/(total_com_cat+total_sem_cat)*100):>5.1f}%)")
print(f"Total de Acidentes.............: {total_com_cat + total_sem_cat:>10,}")
print(f"{'='*70}")
print(f"Taxa Média de Registro CAT.....: {cat_por_ano['taxa_registro'].mean():>9.1f}%")
print(f"Ano com Maior Taxa.............: {cat_por_ano.loc[cat_por_ano['taxa_registro'].idxmax(), 'ano']:.0f} ({cat_por_ano['taxa_registro'].max():.1f}%)")
print(f"Ano com Menor Taxa.............: {cat_por_ano.loc[cat_por_ano['taxa_registro'].idxmin(), 'ano']:.0f} ({cat_por_ano['taxa_registro'].min():.1f}%)")
print(f"{'='*70}\n")
```

### 9.5. Análise de Correlação entre Variáveis

```python
# Seleciona apenas as colunas numéricas para análise de correlação
colunas_analise = [
    'total_acidentes', 
    'com_cat_registrada',
    'acidentes_tipicos', 
    'acidentes_trajeto', 
    'doenca_trabalho',
    'sem_cat_registrada'
]

# Calcula a matriz de correlação
correlacao = df_petroquimico[colunas_analise].corr()

# Cria a figura com dois subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

# 1. Heatmap de correlação
sns.heatmap(correlacao, annot=True, fmt='.2f', cmap='RdYlGn', center=0,
            square=True, linewidths=2, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1, ax=ax1, annot_kws={'size': 11, 'weight': 'bold'})

ax1.set_title('Matriz de Correlação entre Tipos de Acidentes', 
              fontsize=14, fontweight='bold', pad=20)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right', fontsize=10)
ax1.set_yticklabels(ax1.get_yticklabels(), rotation=0, fontsize=10)

# 2. Gráfico de dispersão - Relação entre variáveis mais correlacionadas
# Encontra as duas variáveis com maior correlação (excluindo diagonal e total_acidentes)
correlacao_sem_diagonal = correlacao.copy()
np.fill_diagonal(correlacao_sem_diagonal.values, 0)

# Remove linhas/colunas de total_acidentes para análise
correlacao_filtrada = correlacao_sem_diagonal.drop('total_acidentes', axis=0).drop('total_acidentes', axis=1)

# Encontra o par mais correlacionado
max_corr_idx = correlacao_filtrada.abs().unstack().idxmax()
var1, var2 = max_corr_idx
valor_corr = correlacao.loc[var1, var2]

# Cria scatter plot
ax2.scatter(df_petroquimico[var1], df_petroquimico[var2], 
            alpha=0.6, s=100, c='#1f77b4', edgecolors='black', linewidth=1)

# Adiciona linha de tendência
z = np.polyfit(df_petroquimico[var1], df_petroquimico[var2], 1)
p = np.poly1d(z)
ax2.plot(df_petroquimico[var1].sort_values(), p(df_petroquimico[var1].sort_values()), 
         "r--", linewidth=2, label=f'Tendência Linear')

ax2.set_xlabel(var1.replace('_', ' ').title(), fontsize=12, fontweight='bold')
ax2.set_ylabel(var2.replace('_', ' ').title(), fontsize=12, fontweight='bold')
ax2.set_title(f'Relação entre {var1.replace("_", " ").title()} e {var2.replace("_", " ").title()}\n(Correlação: {valor_corr:.3f})', 
              fontsize=13, fontweight='bold', pad=20)
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=10)

plt.tight_layout()
plt.show()

# Exibe as correlações mais fortes
print(f"\n{'='*80}")
print(f"ANÁLISE DE CORRELAÇÃO ENTRE VARIÁVEIS")
print(f"{'='*80}")
print(f"\nCorrelações mais fortes (valores absolutos > 0.7):\n")

# Extrai pares de correlação
correlacoes_fortes = []
for i in range(len(correlacao.columns)):
    for j in range(i+1, len(correlacao.columns)):
        if abs(correlacao.iloc[i, j]) > 0.7:
            correlacoes_fortes.append({
                'var1': correlacao.columns[i],
                'var2': correlacao.columns[j],
                'corr': correlacao.iloc[i, j]
            })

# Ordena por valor absoluto
correlacoes_fortes = sorted(correlacoes_fortes, key=lambda x: abs(x['corr']), reverse=True)

for item in correlacoes_fortes:
    tipo = "Positiva" if item['corr'] > 0 else "Negativa"
    print(f"  {item['var1']:<25} <-> {item['var2']:<25} : {item['corr']:>6.3f} ({tipo})")

if not correlacoes_fortes:
    print("  Nenhuma correlação forte encontrada (|r| > 0.7)")

print(f"\n{'='*80}\n")
```
