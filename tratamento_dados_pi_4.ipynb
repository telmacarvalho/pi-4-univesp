## Campos das tabelas de dados
- cnae = Classificação Nacional de Atividades Econômicas

- trj_xx = Acidentes de trajeto no ano 20xx
- doe_xx = Doença do trabalho no ano 20xx

## 2. Unificação dos Dados

```python
# Lista de todos os DataFrames
df = {
    2014: df_2014, 2015: df_2015, 2016: df_2016, 2017: df_2017,
    2018: df_2018, 2019: df_2019, 2020: df_2020, 2021: df_2021,
    2022: df_2022, 2023: df_2023
}
```

```python
# Lista para armazenar os dados processados de cada ano
yearly_data = []

for year, df in df.items():
    # Extrai o sufixo do ano (ex: 14, 15, 23)
    year_suffix = str(year)[-2:]

    # Seleciona apenas as colunas relevantes (cnae e aquelas que terminam com o sufixo do ano)
    cols_to_keep = ['cnae'] + [col for col in df.columns if col.endswith(f'_{year_suffix}')]

    df_year = df[cols_to_keep].copy()

    # Renomeia as colunas
    new_column_names = {
        'cnae': 'cnae',
        f'tot_{year_suffix}': 'total_acidentes',
        f'tcat_{year_suffix}': 'com_cat_registrada',
        f'tip_{year_suffix}': 'acidentes_tipicos',
        f'trj_{year_suffix}': 'acidentes_trajeto',
        f'doe_{year_suffix}': 'doenca_trabalho',
        f'scat_{year_suffix}': 'sem_cat_registrada'
    }
    df_year.rename(columns=new_column_names, inplace=True)

    # Adiciona a coluna 'ano'
    df_year['ano'] = year

    yearly_data.append(df_year)

# Concatena todos os dataframes anuais em um só
df_consolidado = pd.concat(yearly_data, ignore_index=True)

print("Formato do DataFrame consolidado:", df_consolidado.shape)
df_consolidado.head()
```

## 4. Limpeza e Conversão de Tipos

```python
# Seleciona as colunas de contagem de acidentes
colunas_numericas = [
    'total_acidentes', 'com_cat_registrada', 'acidentes_tipicos',
    'acidentes_trajeto', 'doenca_trabalho', 'sem_cat_registrada'
]

# Limpa e converte
for col in colunas_numericas:
    # Garante que a coluna seja do tipo string para usar o .str
    df_consolidado[col] = df_consolidado[col].astype(str)

    # Substitui os caracteres não-numéricos
    df_consolidado[col] = df_consolidado[col].str.replace('.', '', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('–', '0', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('-', '0', regex=False)

    # Substitui a string 'nan'
    df_consolidado[col] = df_consolidado[col].str.replace('nan', '0', regex=False)

    # Converte para numérico, preenchendo valores nulos com 0
    df_consolidado[col] = pd.to_numeric(df_consolidado[col]).fillna(0).astype(int)

# Garante que 'cnae' seja tratado como string
df_consolidado['cnae'] = df_consolidado['cnae'].astype(str)

# Tipo dos dados do novo df
print(df_consolidado.info())
```

## 5. Filtragem do Setor Petroquímico
CNAE : ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

```python
# Define os CNAEs específicos (primeiros 4 dígitos)
cnaes_especificos = ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

# Filtra os registros CNAE que começam com os códigos específicos
df_petroquimico = df_consolidado[df_consolidado['cnae'].isin(cnaes_especificos)].copy().reset_index(drop=True)
df_petroquimico.head()
```

```python
# Ordena os dados por cnae e ano
df_petroquimico = df_petroquimico.sort_values(by=['cnae', 'ano']).reset_index(drop=True)
df_petroquimico.head()
```

## 6. Feature Engineering
Definição de Múltiplos Targets

```python
# Define a lista de colunas que queremos prever
colunas_alvo = [
    'total_acidentes',
    'com_cat_registrada',
    'acidentes_tipicos',
    'acidentes_trajeto',
    'doenca_trabalho',
    'sem_cat_registrada'
]

# Cria um df temporário com os targets
df_alvos = df_petroquimico[['cnae'] + colunas_alvo]
df_alvos.head()
```

```python
# Cria outro df com os valores das colunas do ano seguinte para serem os targets
y_multi_alvo = df_alvos.groupby('cnae').shift(-1)
y_multi_alvo.head()
```

```python
# Concatena os dois dfs e remove linhas com NaN
df_final_multi = pd.concat([df_petroquimico, y_multi_alvo.add_suffix('_seguinte')], axis=1)
df_final_multi = df_final_multi.dropna()
df_final_multi.head()
```

## 9. Visualização de Dados
Análise gráfica dos acidentes de trabalho no setor petroquímico

```python
# Importação das bibliotecas de visualização
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurações de estilo
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("Bibliotecas de visualização importadas com sucesso!")
```

### 9.1. Evolução Temporal dos Acidentes por Ano

```python
# Agrupa dados por ano e soma os acidentes
acidentes_por_ano = df_petroquimico.groupby('ano').agg({
    'total_acidentes': 'sum',
    'acidentes_tipicos': 'sum',
    'acidentes_trajeto': 'sum',
    'doenca_trabalho': 'sum'
}).reset_index()

# Cria o gráfico
fig, ax = plt.subplots(figsize=(14, 7))

# Plota as linhas
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['total_acidentes'],
        marker='o', linewidth=2.5, markersize=8, label='Total de Acidentes', color='#d62728')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['acidentes_tipicos'],
        marker='s', linewidth=2, markersize=7, label='Acidentes Típicos', color='#1f77b4')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['acidentes_trajeto'],
        marker='^', linewidth=2, markersize=7, label='Acidentes de Trajeto', color='#ff7f0e')
ax.plot(acidentes_por_ano['ano'], acidentes_por_ano['doenca_trabalho'],
        marker='D', linewidth=2, markersize=7, label='Doenças do Trabalho', color='#2ca02c')

# Personalização
ax.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax.set_ylabel('Número de Acidentes', fontsize=12, fontweight='bold')
ax.set_title('Evolução Temporal dos Acidentes de Trabalho no Setor Petroquímico (2014-2023)',
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='best', fontsize=10, framealpha=0.9)
ax.grid(True, alpha=0.3)
ax.set_xticks(acidentes_por_ano['ano'])

# Adiciona valores no topo de cada ponto
for i, row in acidentes_por_ano.iterrows():
    ax.text(row['ano'], row['total_acidentes'] + 300, f"{int(row['total_acidentes']):,}",
            ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```

### 9.2. Distribuição por Tipo de Acidente

```python
# Calcula a distribuição total por tipo de acidente
distribuicao_tipos = {
    'Acidentes Típicos': df_petroquimico['acidentes_tipicos'].sum(),
    'Acidentes de Trajeto': df_petroquimico['acidentes_trajeto'].sum(),
    'Doenças do Trabalho': df_petroquimico['doenca_trabalho'].sum()
}

# Cria subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico de Pizza
cores = ['#1f77b4', '#ff7f0e', '#2ca02c']
wedges, texts, autotexts = ax1.pie(distribuicao_tipos.values(),
                                     labels=distribuicao_tipos.keys(),
                                     autopct='%1.1f%%',
                                     colors=cores,
                                     startangle=90,
                                     explode=(0.05, 0.05, 0.05),
                                     shadow=True)

# Personaliza os textos do gráfico de pizza
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(11)
    autotext.set_fontweight('bold')

for text in texts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

ax1.set_title('Distribuição Proporcional dos Tipos de Acidentes\n(2014-2023)',
              fontsize=13, fontweight='bold', pad=20)

# Gráfico de Barras Horizontais
tipos = list(distribuicao_tipos.keys())
valores = list(distribuicao_tipos.values())
barras = ax2.barh(tipos, valores, color=cores, edgecolor='black', linewidth=1.5)

# Adiciona valores nas barras
for i, (bar, valor) in enumerate(zip(barras, valores)):
    ax2.text(valor + 500, i, f'{valor:,}', va='center', fontsize=11, fontweight='bold')

ax2.set_xlabel('Número Total de Acidentes', fontsize=12, fontweight='bold')
ax2.set_title('Quantidade Total por Tipo de Acidente\n(2014-2023)',
              fontsize=13, fontweight='bold', pad=20)
ax2.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```

### 9.3. Comparativo de Acidentes por CNAE

```python
# Dicionário com descrições dos CNAEs
cnae_descricoes = {
    '0910': 'Extração de Petróleo e Gás',
    '1921': 'Fabricação de Produtos Petroquímicos',
    '2513': 'Embalagens Metálicas',
    '3321': 'Instalação de Máquinas',
    '3520': 'Distribuição de Gás',
    '4223': 'Obras de Infraestrutura',
    '4292': 'Montagem de Estruturas',
    '4321': 'Instalações Elétricas',
    '4940': 'Transporte Dutoviário',
    '7112': 'Serviços de Engenharia'
}

# Agrupa por CNAE
acidentes_por_cnae = df_petroquimico.groupby('cnae').agg({
    'total_acidentes': 'sum',
    'acidentes_tipicos': 'sum',
    'acidentes_trajeto': 'sum',
    'doenca_trabalho': 'sum'
}).reset_index()

# Ordena por total de acidentes
acidentes_por_cnae = acidentes_por_cnae.sort_values('total_acidentes', ascending=True)

# Adiciona descrições
acidentes_por_cnae['descricao'] = acidentes_por_cnae['cnae'].map(cnae_descricoes)
acidentes_por_cnae['label'] = acidentes_por_cnae['cnae'] + ' - ' + acidentes_por_cnae['descricao']

# Cria o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(14, 8))

y_pos = np.arange(len(acidentes_por_cnae))
width = 0.8

# Barras empilhadas
p1 = ax.barh(y_pos, acidentes_por_cnae['acidentes_tipicos'], width,
             label='Acidentes Típicos', color='#1f77b4', edgecolor='black', linewidth=0.5)
p2 = ax.barh(y_pos, acidentes_por_cnae['acidentes_trajeto'], width,
             left=acidentes_por_cnae['acidentes_tipicos'],
             label='Acidentes de Trajeto', color='#ff7f0e', edgecolor='black', linewidth=0.5)
p3 = ax.barh(y_pos, acidentes_por_cnae['doenca_trabalho'], width,
             left=acidentes_por_cnae['acidentes_tipicos'] + acidentes_por_cnae['acidentes_trajeto'],
             label='Doenças do Trabalho', color='#2ca02c', edgecolor='black', linewidth=0.5)

# Adiciona o total no final de cada barra
for i, (idx, row) in enumerate(acidentes_por_cnae.iterrows()):
    ax.text(row['total_acidentes'] + 500, i, f"{int(row['total_acidentes']):,}",
            va='center', fontsize=10, fontweight='bold')

# Personalização
ax.set_yticks(y_pos)
ax.set_yticklabels(acidentes_por_cnae['label'], fontsize=10)
ax.set_xlabel('Número de Acidentes', fontsize=12, fontweight='bold')
ax.set_title('Comparativo de Acidentes por CNAE no Setor Petroquímico (2014-2023)',
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='lower right', fontsize=10, framealpha=0.9)
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```

### 9.4. Tendências Temporais por CNAE

```python
# Dicionário com descrições abreviadas dos CNAEs
cnae_desc_curtas = {
    '0910': 'Extração Petróleo/Gás',
    '1921': 'Fab. Petroquímicos',
    '2513': 'Embalagens Metálicas',
    '3321': 'Instalação Máquinas',
    '3520': 'Distribuição Gás',
    '4223': 'Obras Infraestrutura',
    '4292': 'Montagem Estruturas',
    '4321': 'Instalações Elétricas',
    '4940': 'Transporte Dutoviário',
    '7112': 'Serv. Engenharia'
}

# Cria o gráfico
fig, ax = plt.subplots(figsize=(16, 9))

# Define paleta de cores
cores_cnae = plt.cm.tab10(np.linspace(0, 1, 10))

# Plota cada CNAE
for i, cnae in enumerate(df_petroquimico['cnae'].unique()):
    dados_cnae = df_petroquimico[df_petroquimico['cnae'] == cnae]
    ax.plot(dados_cnae['ano'], dados_cnae['total_acidentes'],
            marker='o', linewidth=2, markersize=7,
            label=f"{cnae} - {cnae_desc_curtas[cnae]}",
            color=cores_cnae[i], alpha=0.8)

# Personalização
ax.set_xlabel('Ano', fontsize=12, fontweight='bold')
ax.set_ylabel('Total de Acidentes', fontsize=12, fontweight='bold')
ax.set_title('Evolução Temporal dos Acidentes por CNAE (2014-2023)',
             fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='upper left', fontsize=9, ncol=2, framealpha=0.9)
ax.grid(True, alpha=0.3)
ax.set_xticks(df_petroquimico['ano'].unique())

plt.tight_layout()
plt.show()

plt.tight_layout()
plt.show()
```
