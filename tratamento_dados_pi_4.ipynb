Quantidade de acidentes do trabalho, por situação do registro e motivo, segundo a Classificação Nacional de Atividades Econômicas (CNAE), no Brasil.

## Campos das tabelas de dados
- cnae = Classificação Nacional de Atividades Econômicas
- tot_xx = Total geral de acidentes no trabalho no ano 20xx

- tip_xx = Acidentes típicos no ano 20xx
- trj_xx = Acidentes de trajeto no ano 20xx
- doe_xx = Doença do trabalho no ano 20xx

## 2. Unificação dos Dados

```python
# Lista de todos os DataFrames
df = {
    2014: df_2014, 2015: df_2015, 2016: df_2016, 2017: df_2017,
    2018: df_2018, 2019: df_2019, 2020: df_2020, 2021: df_2021,
    2022: df_2022, 2023: df_2023
}
```

```python
# Lista para armazenar os dados processados de cada ano
yearly_data = []

for year, df in df.items():
    # Extrai o sufixo do ano (ex: 14, 15, 23)
    year_suffix = str(year)[-2:]

    # Seleciona apenas as colunas relevantes (cnae e aquelas que terminam com o sufixo do ano)
    cols_to_keep = ['cnae'] + [col for col in df.columns if col.endswith(f'_{year_suffix}')]

    df_year = df[cols_to_keep].copy()

    # Renomeia as colunas
    new_column_names = {
        'cnae': 'cnae',
        f'tot_{year_suffix}': 'total_acidentes',
        f'tcat_{year_suffix}': 'com_cat_registrada',
        f'tip_{year_suffix}': 'acidentes_tipicos',
        f'trj_{year_suffix}': 'acidentes_trajeto',
        f'doe_{year_suffix}': 'doenca_trabalho',
        f'scat_{year_suffix}': 'sem_cat_registrada'
    }
    df_year.rename(columns=new_column_names, inplace=True)

    # Adiciona a coluna 'ano'
    df_year['ano'] = year

    yearly_data.append(df_year)

# Concatena todos os dataframes anuais em um só
df_consolidado = pd.concat(yearly_data, ignore_index=True)

print("Formato do DataFrame consolidado:", df_consolidado.shape)
df_consolidado.head()
```

## 4. Limpeza e Conversão de Tipos

```python
# Seleciona as colunas de contagem de acidentes
colunas_numericas = [
    'total_acidentes', 'com_cat_registrada', 'acidentes_tipicos',
    'acidentes_trajeto', 'doenca_trabalho', 'sem_cat_registrada'
]

# Limpa e converte
for col in colunas_numericas:
    # Garante que a coluna seja do tipo string para usar o .str
    df_consolidado[col] = df_consolidado[col].astype(str)

    # Substitui os caracteres não-numéricos
    df_consolidado[col] = df_consolidado[col].str.replace('.', '', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('–', '0', regex=False)
    df_consolidado[col] = df_consolidado[col].str.replace('-', '0', regex=False)

    # Substitui a string 'nan'
    df_consolidado[col] = df_consolidado[col].str.replace('nan', '0', regex=False)

    # Converte para numérico, preenchendo valores nulos com 0
    df_consolidado[col] = pd.to_numeric(df_consolidado[col]).fillna(0).astype(int)

# Garante que 'cnae' seja tratado como string
df_consolidado['cnae'] = df_consolidado['cnae'].astype(str)

# Tipo dos dados do novo df
print(df_consolidado.info())
```

## 5. Filtragem do Setor Petroquímico
CNAE : ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

```python
# Define os CNAEs específicos (primeiros 4 dígitos)
cnaes_especificos = ['4292', '4940', '3520', '1921', '0910', '3321', '4321', '7112', '4223', '2513']

# Filtra os registros CNAE que começam com os códigos específicos
df_petroquimico = df_consolidado[df_consolidado['cnae'].isin(cnaes_especificos)].copy().reset_index(drop=True)
df_petroquimico.head()
```

```python
# Ordena os dados por cnae e ano
df_petroquimico = df_petroquimico.sort_values(by=['cnae', 'ano']).reset_index(drop=True)
df_petroquimico.head()
```

```python
print(f"Registros encontrados para o setor petroquímico: {df_petroquimico.shape[0]} ")
print(f"Número de CNAEs únicos no setor: {df_petroquimico['cnae'].nunique()}")
```

## 6. Feature Engineering
Definição de Múltiplos Targets

```python
# Define a lista de colunas que queremos prever
colunas_alvo = [
    'total_acidentes',
    'com_cat_registrada',
    'acidentes_tipicos',
    'acidentes_trajeto',
    'doenca_trabalho',
    'sem_cat_registrada'
]

# Cria um df temporário com os targets
df_alvos = df_petroquimico[['cnae'] + colunas_alvo]
df_alvos.head()
```

```python
# Cria outro df com os valores das colunas do ano seguinte para serem os targets
y_multi_alvo = df_alvos.groupby('cnae').shift(-1)
y_multi_alvo.head()
```

```python
# Concatena os dois dfs e remove linhas com NaN
df_final_multi = pd.concat([df_petroquimico, y_multi_alvo.add_suffix('_seguinte')], axis=1)
df_final_multi = df_final_multi.dropna()
df_final_multi.head()
```
